```python
#%matplotlib nbagg
#%matplotlib inline
%matplotlib notebook
#to show matplotlib in jupyter notebook

from ipynb.fs.full.piwebapi import getData

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
from itertools import count
import datetime
```

```python
plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(datetime.datetime.now())
    y_vals.append(getData())
    
    plt.cla()
    #togetridofscientific notation
    #plt.ticklabel_format(axis = 'y' , useOffset=False, style='plain')
    plt.plot(x_vals, y_vals)
    


ani = FuncAnimation(plt.gcf(), animate, interval = 1000)
plt.tight_layout()
plt.show()
```
